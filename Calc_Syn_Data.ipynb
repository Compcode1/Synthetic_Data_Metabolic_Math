#### **Introduction**

Metabolic syndrome is a binary classification system.  The most commonly used criteria for diagnosing metabolic syndrome come from the National Heart, Lung and Blood Institute (NHLBI) and the American Heart Association (AHA). According to these guidelines, a person is diagnosed with metabolic syndrome if they have 3 or more of the following 5 risk factors:
Abdominal obesity: Waist circumference > 40 inches (102 cm) for men or > 35 inches (88 cm) for women
High blood pressure: ≥ 130/85 mmHg or on antihypertensive medication
Elevated fasting blood glucose: ≥ 100 mg/dL or on glucose-lowering medication
High triglycerides: ≥ 150 mg/dL or on lipid-lowering medication
Low HDL cholesterol: < 40 mg/dL for men or < 50 mg/dL for women, or on lipid-lowering medication

The binary nature comes from the fact that a person either meets the criteria for metabolic syndrome (having 3 or more risk factors) or does not (having fewer than 3 risk factors). There is no gradient or spectrum - it's a yes/no diagnosis based on meeting the threshold number of criteria.

The statistical nature of obesity and metabolic risk factors in younger individuals presents a complex picture that evolves over time. Cross-sectional studies often show that younger people with elevated BMI or increased waist circumference have fewer metabolic risk factors compared to older individuals, which can be misleading if not contextualized with longitudinal data. These long-term studies reveal that individuals who are overweight or obese at a younger age are statistically more likely to develop metabolic risk factors as they age. This is reflected in the changing incidence and prevalence rates of metabolic conditions, which are initially lower in younger populations but increase significantly with age, especially among those who were overweight or obese in their youth.

Risk ratios and odds ratios further illustrate this trend, showing that while younger overweight individuals may have a lower relative risk initially, their likelihood of developing conditions like diabetes or hypertension increases over time. Predictive models, including regression analyses and time-to-event analyses, consistently demonstrate that BMI is a significant predictor of future metabolic risk, even when current blood work appears normal. Population attributable risk calculations often indicate that a substantial portion of future metabolic risk in populations can be attributed to early-life obesity.

Survival analyses, such as Kaplan-Meier curves, typically show a steeper decline in the probability of remaining free of metabolic risk factors over time for younger individuals with elevated BMI. Long-term cohort studies, exemplified by research like the Framingham Heart Study, have provided robust evidence that individuals who are overweight or obese in early adulthood face a significantly higher risk of developing metabolic syndrome and related conditions later in life. These statistical findings underscore the importance of early intervention and continuous monitoring, as they consistently demonstrate that while younger individuals with elevated BMI may initially present with fewer metabolic risk factors, their risk increases significantly over time, making early-life obesity a crucial predictor of future metabolic health.

A risk model that incorporates BMI as an independent risk factor acknowledges the statistical reality that elevated BMI, particularly in younger individuals, is a significant predictor of future metabolic health issues, even when current metabolic markers are normal. This approach recognizes that the impact of obesity on health is not always immediately apparent in blood work or other standard metabolic measurements, especially in younger populations. By including BMI as a separate variable, the model captures the potential for future metabolic dysregulation that may develop over time, even in the absence of current abnormalities. This is particularly important given the time-dependent nature of obesity-related health risks, where the cumulative effects of excess weight may take years to manifest clinically. This health risk score model effectively incorporates BMI as an independent risk factor alongside other known metabolic syndrome risk variables. By assigning points based on BMI categories independently of other factors, the model ensures that BMI contributes to the overall risk score even if other metabolic markers are normal, reflecting the potential for future risk development. The graduated scoring system for BMI, with higher points for higher BMI categories, reflects the increasing risk associated with higher levels of obesity. Notably, the maximum score for BMI is equal to the maximum scores for high blood pressure and elevated fasting blood glucose, putting BMI on par with these significant metabolic risk factors. 

The additive nature of the model allows BMI's contribution to be cumulative with other risk factors, enabling a comprehensive risk assessment that considers both current metabolic abnormalities and the potential for future risk indicated by elevated BMI. #### **Metabolic Syndrome Analysis**

This section demonstrates advanced data manipulation techniques using pandas to analyze metabolic syndrome in the adult dataset. A custom function, `has_metabolic_syndrome`, is defined to evaluate individual rows against multiple health criteria. This function employs conditional logic to check for various risk factors, including abdominal obesity, elevated triglycerides, reduced HDL cholesterol, high blood pressure, and elevated fasting glucose. The `apply` method is then used to efficiently process the entire DataFrame, creating a new column that indicates the presence of metabolic syndrome for each individual.  The analysis concludes by calculating the percentage of individuals with metabolic syndrome using two methods: one based on the calculated criteria and another using an existing column in the dataset. The use of the `mean()` function on boolean columns provides a quick way to compute percentages. The results show perfect agreement between the calculated and actual percentages, validating the accuracy of the custom function and demonstrating the effectiveness of the implemented data analysis techniques.

The prevalence of metabolic syndrome in our synthetic dataset, at 34.67%, aligns closely with recent estimates for the adult US population based on National Health and Nutrition Examination Survey (NHANES) data. According to studies using NHANES data, the prevalence of metabolic syndrome in US adults has been increasing over the past few decades. Recent estimates suggest that approximately 34-35% of US adults meet the criteria for metabolic syndrome, which is remarkably consistent with our dataset's 34.67% prevalence. This similarity lends credibility to our synthetic dataset as a reasonable representation of real-world trends. 

```python
import pandas as pd

# Load the dataset
file_path = 'df_adult.csv'
df_adult = pd.read_csv(file_path)

# Define risk factors
def has_metabolic_syndrome(row):
    criteria = 0
    
    # Abdominal Obesity
    if (row['Gender'] == 'Male' and row['Waist_Circumference'] > 40) or (row['Gender'] == 'Female' and row['Waist_Circumference'] > 35):
        criteria += 1
    
    # Elevated Triglycerides
    if row['Triglycerides'] >= 150:
        criteria += 1
    
    # Reduced HDL Cholesterol
    if (row['Gender'] == 'Male' and row['HDL_Cholesterol'] < 40) or (row['Gender'] == 'Female' and row['HDL_Cholesterol'] < 50):
        criteria += 1
    
    # Elevated Blood Pressure
    if row['High_Blood_Pressure'] == 1:
        criteria += 1
    
    # Elevated Fasting Glucose
    if row['Fasting_Blood_Glucose'] >= 100:
        criteria += 1
    
    return criteria >= 3

# Calculate the percentage of individuals with metabolic syndrome based on criteria
df_adult['Calculated_Metabolic_Syndrome'] = df_adult.apply(has_metabolic_syndrome, axis=1)
calculated_metabolic_syndrome_percentage = df_adult['Calculated_Metabolic_Syndrome'].mean() * 100

# Calculate the percentage of individuals with metabolic syndrome based on the 'Metabolic_Syndrome' column
actual_metabolic_syndrome_percentage = df_adult['Metabolic_Syndrome'].mean() * 100

# Print the results
print(f"Percentage of individuals with metabolic syndrome (calculated): {calculated_metabolic_syndrome_percentage:.2f}%")
print(f"Percentage of individuals with metabolic syndrome (actual): {actual_metabolic_syndrome_percentage:.2f}%")
```

Notably, this system incorporates BMI as a standalone variable, separate from the metabolic syndrome criteria. The scoring system weighs different health factors based on their perceived impact:

BMI: Scores range from 0 to 3, with 3 points for obesity (BMI ≥ 30) and 2 points for overweight (25 ≤ BMI < 30). A normal BMI (18.5-24.9) receives no points, while being underweight adds 1 point.

Blood Pressure: High blood pressure adds 3 points.

The distribution of risk levels shifts significantly with age. Conversely, the proportion of individuals in high and very high-risk categories increases with age, with the 70+ group showing the highest percentages (32.37% and 18.86% respectively). The moderate risk category remains relatively stable across all age groups, consistently representing the largest proportion of individuals (ranging from 24.71% to 32.72%).

The risk score model provides a more nuanced and comprehensive assessment of metabolic health risks across all BMI categories, offering significant advantages over the binary Metabolic Syndrome classification, particularly for younger individuals. This approach reveals important trends in metabolic health that might be overlooked by traditional methods. For instance, while the Normal (18.5-24.9 BMI) and Overweight (25-29.9 BMI) categories show lower Metabolic Syndrome prevalence compared to the Obese category (BMI ≥ 30), the risk score calculations indicate substantial metabolic risk even at these lower BMI levels. This is supported by recent studies showing that overweight individuals are 2.4 times more likely to develop Metabolic Syndrome compared to those with normal BMI, while obese individuals face a 4.4 times higher risk. 

The model's ability to assign risk scores across a continuous spectrum allows for early identification of potential metabolic issues in Normal and Overweight individuals, enabling preventive interventions before progression to obesity or full-blown Metabolic Syndrome.  It also identifies high risk in the Underweight category (BMI < 18.5), where individuals may have a lower prevalence of Metabolic Syndrome but still face significant health risks due to other metabolic abnormalities. By providing this more detailed and personalized risk assessment, the model serves as a powerful tool for early intervention, potentially improving long-term health outcomes by addressing risks before they manifest as clinical Metabolic Syndrome, especially in younger populations where traditional risk factors may not yet be apparent.




```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'df_adult.csv'
df_adult = pd.read_csv(file_path)

# Define age categories
def categorize_age(age):
    if 18 <= age <= 29:
        return "18-29"
    elif 30 <= age <= 39:
        return "30-39"
    elif 40 <= age <= 49:
        return "40-49"
    elif 50 <= age <= 59:
        return "50-59"
    elif 60 <= age <= 69:
        return "60-69"
    else:
        return "70+"

# Add age category to the dataframe
df_adult['Age_Category'] = df_adult['Age'].apply(categorize_age)

# Define metabolic syndrome function
def has_metabolic_syndrome(row):
    criteria = 0
    if (row['Gender'] == 'Male' and row['Waist_Circumference'] > 40) or (row['Gender'] == 'Female' and row['Waist_Circumference'] > 35):
        criteria += 1
    if row['Triglycerides'] >= 150:
        criteria += 1
    if (row['Gender'] == 'Male' and row['HDL_Cholesterol'] < 40) or (row['Gender'] == 'Female' and row['HDL_Cholesterol'] < 50):
        criteria += 1
    if row['High_Blood_Pressure'] == 1:
        criteria += 1
    if row['Fasting_Blood_Glucose'] >= 100:
        criteria += 1
    return criteria >= 3

# Calculate metabolic syndrome
df_adult['Calculated_Metabolic_Syndrome'] = df_adult.apply(has_metabolic_syndrome, axis=1)

def calculate_health_risk_score(row):
    score = 0
    
    # BMI
    if row['BMI'] >= 30:
        score += 3
    elif 25 <= row['BMI'] < 30:
        score += 2
    elif 18.5 <= row['BMI'] < 25:
        score += 0
    else:
        score += 1

    # Blood Pressure
    if row['High_Blood_Pressure'] == 1:
        score += 3

    # Cholesterol
    if row['HDL_Cholesterol'] < 40:
        score += 2
    elif 40 <= row['HDL_Cholesterol'] < 60:
        score += 1

    # Triglycerides
    if row['Triglycerides'] >= 200:
        score += 2
    elif 150 <= row['Triglycerides'] < 200:
        score += 1

    # Fasting Blood Glucose
    if row['Fasting_Blood_Glucose'] >= 126:
        score += 3
    elif 100 <= row['Fasting_Blood_Glucose'] < 126:
        score += 2

    return score

def categorize_risk(score):
    if pd.isna(score):
        return "No Risk"
    score = float(score)
    if score == 0:
        return "No Risk"
    elif 1 <= score <= 2:
        return "low"
    elif 3 <= score <= 5:
        return "moderate"
    elif 6 <= score <= 8:
        return "high"
    elif score > 9:
        return "very high"
    return "No Risk"

# Apply the health risk score calculation to each row
df_adult['Health_Risk_Score'] = df_adult.apply(calculate_health_risk_score, axis=1)

# Categorize the risk based on the score
df_adult['Risk_Category'] = df_adult['Health_Risk_Score'].apply(categorize_risk)

# Calculate percentages for each age category
age_categories = df_adult['Age_Category'].unique()
results = []

for category in age_categories:
    category_df = df_adult[df_adult['Age_Category'] == category]
    total_count = len(category_df)
    
    metabolic_syndrome_pct = (category_df['Calculated_Metabolic_Syndrome'].sum() / total_count) * 100
    
    risk_percentages = category_df['Risk_Category'].value_counts(normalize=True) * 100
    
    results.append({
        'Age_Category': category,
        'Metabolic_Syndrome_%': metabolic_syndrome_pct,
        'No_Risk_%': risk_percentages.get('No Risk', 0),
        'Low_Risk_%': risk_percentages.get('low', 0),
        'Moderate_Risk_%': risk_percentages.get('moderate', 0),
        'High_Risk_%': risk_percentages.get('high', 0),
        'Very_High_Risk_%': risk_percentages.get('very high', 0)
    })

# Create a dataframe with the results
result_df = pd.DataFrame(results)

# Sort the dataframe by age category
result_df['Age_Category'] = pd.Categorical(result_df['Age_Category'], 
                                           categories=["18-29", "30-39", "40-49", "50-59", "60-69", "70+"], 
                                           ordered=True)
result_df = result_df.sort_values('Age_Category')

print("\nFinal results:")
print(result_df.to_string(index=False))
```

By incorporating BMI as a significant factor, our model can identify potential health risks in younger, overweight, or obese individuals who might not yet meet the criteria for metabolic syndrome but are at increased risk for future metabolic complications.

In conclusion, these data strongly support the premise of our model, demonstrating that BMI is indeed a more significant variable than age in determining metabolic health risk. This aligns with known statistical trends and provides a more nuanced and potentially more accurate assessment of individual health risks across all age groups.


```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'df_adult.csv'
df_adult = pd.read_csv(file_path)

# Define age categories
def categorize_age(age):
    if 18 <= age <= 29:
        return "18-29"
    elif 30 <= age <= 39:
        return "30-39"
    elif 40 <= age <= 49:
        return "40-49"
    elif 50 <= age <= 59:
        return "50-59"
    elif 60 <= age <= 69:
        return "60-69"
    else:
        return "70+"

# Define BMI categories
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Define health risk score calculation
def calculate_health_risk_score(row):
    score = 0
    
    # BMI
    if row['BMI'] >= 30:
        score += 3
    elif 25 <= row['BMI'] < 30:
        score += 2
    elif 18.5 <= row['BMI'] < 25:
        score += 0
    else:
        score += 1

    # Blood Pressure
    if row['High_Blood_Pressure'] == 1:
        score += 3

    # Cholesterol
    if row['HDL_Cholesterol'] < 40:
        score += 2
    elif 40 <= row['HDL_Cholesterol'] < 60:
        score += 1

    # Triglycerides
    if row['Triglycerides'] >= 200:
        score += 2
    elif 150 <= row['Triglycerides'] < 200:
        score += 1

    # Fasting Blood Glucose
    if row['Fasting_Blood_Glucose'] >= 126:
        score += 3
    elif 100 <= row['Fasting_Blood_Glucose'] < 126:
        score += 2

    return score

# Add age and BMI categories to the dataframe
df_adult['Age_Category'] = df_adult['Age'].apply(categorize_age)
df_adult['BMI_Category'] = df_adult['BMI'].apply(categorize_bmi)

# Calculate health risk score
df_adult['Health_Risk_Score'] = df_adult.apply(calculate_health_risk_score, axis=1)

# Calculate the converted risk score
df_adult['Converted_Risk_Score'] = (df_adult['Health_Risk_Score'] / 18 * 100).round().astype(int)

# Calculate average risk score per age group
avg_risk_by_age = df_adult.groupby('Age_Category')['Converted_Risk_Score'].mean().round(2)

# Calculate average risk score per BMI category
avg_risk_by_bmi = df_adult.groupby('BMI_Category')['Converted_Risk_Score'].mean().round(2)

# Sort the BMI categories in a logical order
bmi_order = ['Underweight', 'Normal', 'Overweight', 'Obese']
avg_risk_by_bmi = avg_risk_by_bmi.reindex(bmi_order)

# Print results
print("Average Risk Score by Age Group:")
print(avg_risk_by_age)
print("\nAverage Risk Score by BMI Category:")
print(avg_risk_by_bmi)
```
